# c. Los 10 puntos más cercanos geográficamente a cada dato faltante usando las coordenadas.

- Obtener los alquileres que no tienen precio
- Por cada alquiler que no tiene precio obtenido:
   1. calculo la distancia al resto de los
   2.alquileres con precio
   3. ordeno la lista en orden ascendente de distancia
   4. tomo los 10 primeros
   5. calculo el promedio de los puntos cercanos
   6. inputo el valor al registro

```python
alquileres_sin_precio = data.loc[data['price'].isna()].copy()
alquileres_sin_precio.head()
```

```python
alquileres_con_precio = data.loc[data['price'].notnull()].copy()
alquileres_con_precio.head()
```

```python
import math

def distancia_entre_puntos(x1, y1, x2, y2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
```

```python
#Por cada alquiler que no tiene precio obtenido:

for index, row in alquileres_sin_precio.iterrows():

  # Calculo la distancia al resto de los alquileres con precio
  punto_base = (row['latitude'], row['longitude'])
  alquileres_con_precio['distancia'] = alquileres_con_precio.apply(lambda fila: distancia_entre_puntos(punto_base[0], punto_base[1], fila['latitude'], fila['longitude']), axis=1)

  # Ordeno la lista en orden ascendente de distancia
  df_ordenado = alquileres_con_precio.sort_values('distancia')
  # tomo los 10 primeros y calculo el promedio de los puntos cercanos
  promedio = df_ordenado.head(10)['price'].mean()
  # inputo el valor al registro actual
  alquileres_sin_precio.loc[index, 'price'] = promedio
```
