Utilizando los archivos conicet_personas_2020.xlsx , conicet_ref_sexo.xlsx y conicet_ref_grado_academico.xlsx
disponibles en el dataset de la unidad 2, genere una tabla en la cual se informe cuántos empleados
de CONICET hay de cada sexo para cada máximo grado académico en 2020.

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
archivo= '/content/drive/MyDrive/TUIA/Fundamentos CD/datasets/unidad2/conicet_personas_2020.xlsx'
conicet_personas = pd.read_excel(archivo)
conicet_personas.head()
```

```python
archivo= '/content/drive/MyDrive/TUIA/Fundamentos CD/datasets/unidad2/conicet_ref_sexo.xlsx'
conicet_sexo = pd.read_excel(archivo)
conicet_sexo.head()
```

```python
archivo= '/content/drive/MyDrive/TUIA/Fundamentos CD/datasets/unidad2/conicet_ref_grado_academico.xlsx'
conicet_grado = pd.read_excel(archivo)
conicet_grado.head()
```

```python
# unir dataframe personas con dataframe sexo
tabla_con_sexo = pd.merge(conicet_personas, conicet_sexo, on='sexo_id', how='outer')
tabla_con_sexo.head()
```

```python
# unir dataframe tabla_con_sexo con dataframe grado academico
tabla_con_sexo_grado = pd.merge(tabla_con_sexo, conicet_grado, left_on='maximo_grado_academico_id', right_on='grado_academico_id', how='outer')
tabla_con_sexo_grado.rename(columns={"descripcion":"grado_academico","sexo_descripcion":"sexo"},inplace=True)
tabla_con_sexo_grado.head()

```

```python
# generar una tabla totalizadora por sexo y por maximo grado academico
(tabla_con_sexo_grado.groupby(["sexo","grado_academico"])).count()['persona_id']
```

```python
#reorganizo la tabla para mas claridad
tabla_pivot = tabla_con_sexo_grado.pivot_table(index='grado_academico',
                                               columns='sexo',
                                               values='persona_id',
                                               aggfunc='count')
print(tabla_pivot)
```
