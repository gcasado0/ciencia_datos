Ejercicio N°8
Utilizando el archivo incendios-cantidad-causas-provincia_2022.csv del Ejercicio N° 2 generar una tabla
que muestre el número de incendios intencionales por provincia para cada año (por ej.: las
provincias como filas y los años en las columnas).

```python
import pandas as pd
import numpy as np
from google.colab import drive
drive.mount('/content/drive')
```

```python
archivo= '/content/drive/MyDrive/TUIA/Fundamentos CD/datasets/unidad2/incendios-cantidad-causas-provincia_2022.csv'
incendios = pd.read_csv(archivo, encoding='latin-1', skiprows=3)
# quito espacios de los nombres de las provincias
incendios["provincia"] = incendios["provincia"].str.strip()
# cordoba esta escrito con acento y sin acento
incendios["provincia"].replace("Cordoba","Córdoba",inplace=True)

incendios.head()
```

# Tabla que muestre el número de incendios intencionales por provincia para cada año
(por ej.: las provincias como filas y los años en las columnas).

```python
tabla_pivot = incendios.pivot_table(index='provincia',
                                               columns='anio',
                                               values='intencional',
                                               aggfunc='sum')
tabla_pivot
```
