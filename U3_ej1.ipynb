````python
sns.boxplot(data=data, x='density', orient='h')
plt.xlabel('density')
plt.show()
````

• ¿Cuál de los dos conjuntos de observaciones (densidad o pH) presenta mayor variabilidad?

````python
#Coeficiente de Variación (CV)
cv_density = data["density"].describe().loc['std']/abs(data["density"].describe().loc['mean'])

cv_pH = data["pH"].describe().loc['std']/abs(data["pH"].describe().loc['mean'])

print(f"Coeficiente de Variación (CV) de densidad: {round(cv_density,4)}")
print(f"Coeficiente de Variación (CV) de pH: {round(cv_pH,4)}")
````

**Conclusión: El pH presenta mayor variabilidad**

# 5. Realice una tabla de frecuencias para resumir la distribución de los vinos del dataset en función del puntaje asignado según su calidad (quality).

* ¿Cuál de los puntajes fue recibido por una mayor cantidad de vinos?
* ¿Qué porcentaje de los vinos de la muestra recibieron la calificación más baja?

````python
data['quality'].value_counts().sort_index()
````

````python
print(f" ¿Cuál de los puntajes fue recibido por una mayor cantidad de vinos? El valor es: {data['quality'].mode()[0]}")
````

````python
#¿Qué porcentaje de los vinos de la muestra recibieron la calificación más baja?
frecuencia_relativa = data['quality'].value_counts().reset_index()
frecuencia_relativa.rename(columns={"count":"Frecuencia_abs"},inplace=True)
frecuencia_relativa.set_index("quality",inplace=True)
frecuencia_relativa['Frecuencia_rel'] = round(frecuencia_relativa['Frecuencia_abs']*100/frecuencia_relativa['Frecuencia_abs'].sum(),2)
frecuencia_relativa
````

¿Qué porcentaje de los vinos de la muestra recibieron la calificación más baja?

El 0.74% de los vinos recibieron la calificacion mas baja.
